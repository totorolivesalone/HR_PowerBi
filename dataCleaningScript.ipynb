```python
#dropping Employee count column as all employees have that field set to 1 and it's given knowledge
#also dropping Over18 as the information can be deduced during run time using Age field
df=df.drop(columns=['EmployeeCount','Over18'])
```

```python
df.head()
```

```python
#median salary calculation from data set for various roles .
df['Role_Median_Salary'] = df.groupby('JobRole')['MonthlyIncome'].transform('median') #broadcast using transform
df.head()
```

```python
# over_worked flag rule : less than two years and working overtime | selected two years because that is when bonuses stock on median can be cashed out
df['flag_overworked']=(df['OverTime']=='Yes') & (df['YearsAtCompany'] <=2)
df.head()
```

```python
# underpaid rule set by pay less than median salary we determined above
df['flag_underpaid']=df['MonthlyIncome']<df['Role_Median_Salary']
df.head()
```

```python
#role satisfaction rule : if user has been in a role for more than three years and job satisfaction score is less than 2
df['flag_stagnate']=(df['YearsInCurrentRole']>3)&(df['JobSatisfaction']<=2)
df.head()
```

```python
#chronic job hopper identifier : worked in more than 4 jobs before here and the 2 year mark is approaching
df['flag_hopper']=(df['NumCompaniesWorked']>=4)&(df['YearsAtCompany']<2)
df.head()
```

```python
#high risk column : if any flags are true in temp dataframe created out of new flags , then we can deem user at high risk
df['High_Risk'] = df[['flag_overworked', 'flag_underpaid',  'flag_stagnate', 'flag_hopper']].any(axis=1)
df.head()
```

## Our Rule-Based Analysis

```python
total_employees=len(df)
total_at_risk=df['High_Risk'].sum()
print("Predicted At Risk :",total_at_risk," employees ,out of ",total_employees," employees")
print("High risk percentage:",((total_at_risk*100)/total_employees))
```

## Moment of Truth : Comparison results between our rules' outcome and reality

```python
comparison=pd.crosstab(df['Attrition'],df['High_Risk'], margins=True)
print("Actual attrition vs High-Risk Flag created:")
print(comparison)
```

```python
actual_quit_rate = df[df['High_Risk'] == True]['Attrition'].value_counts(normalize=True) * 100
print("\nPercentage of your flagged employees who actually quit:")
print(actual_quit_rate)
```
